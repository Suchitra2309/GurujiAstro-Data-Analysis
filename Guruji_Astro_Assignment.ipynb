**Handling Missing Values For Order Data**

```python
# Filling missing values for categorical columns with 'Unknown'
order_data['Astro Status'].fillna('Unknown', inplace=True)
order_data['User Status'].fillna('Unknown', inplace=True)
order_data['Status'].fillna('Unknown', inplace=True)
order_data['Mode'].fillna('Unknown', inplace=True)
order_data['IST Date'].fillna(order_data['IST Date'].mode()[0], inplace=True)  # Filling with mode for IST Date
```

```python
# Converting Consultation Start and End Time columns to datetime format
order_data['Consultation Start Time'] = pd.to_datetime(order_data['Consultation Start Time'], errors='coerce')
order_data['Consultation End Time'] = pd.to_datetime(order_data['Consultation End Time'], errors='coerce')
```

```python
# Filling Consultation Start and End Times with median values
# Filling Consultation Start and End Times with median values
median_start_time = order_data['Consultation Start Time'].median()
median_end_time = order_data['Consultation End Time'].median()
order_data.fillna({'Consultation Start Time': median_start_time}, inplace=True)
order_data.fillna({'Consultation End Time': median_end_time}, inplace=True)
```

```python
# Filling Disconnected By, User Device, and Astro Device with 'Unknown'
order_data.fillna({'Disconnected By': 'Unknown'}, inplace=True)
order_data.fillna({'User Device': 'Unknown'}, inplace=True)
order_data.fillna({'Astro Device': 'Unknown'}, inplace=True)
```

```python
# Dropping IVR RefId, CallChannel, Feedback,goodFeedback and badFeedback due to high missing values
Cleaned_order_data = order_data.drop(columns=['IVR RefId', 'CallChannel','Feedback','goodFeedback','badFeedback'])
```

```python
# Customer Segmentation by Spend
def create_customer_segments(spend):
    if spend == 0:
        return 'Inactive'
    elif spend <= 100:
        return 'Low'
    elif spend <= 500:
        return 'Medium'
    else:
        return 'High'

# Create customer segments
user_segments = Cleaned_users_data.copy()
user_segments['segment'] = user_segments['totalRechargeAmount'].apply(create_customer_segments)
```

```python
# Analyze revenue by segment
segment_revenue = user_segments.groupby('segment')['totalRechargeAmount'].agg(['sum', 'count', 'mean'])
print("Revenue Analysis by Customer Segment:")
print(segment_revenue)
```

```python
# Transaction value analysis
Cleaned_recharge_data['value_segment'] = pd.qcut(Cleaned_recharge_data['Amount'], q=4, labels=['Low', 'Medium-Low', 'Medium-High', 'High'])
transaction_value_analysis = Cleaned_recharge_data.groupby('value_segment')['Amount'].agg(['count', 'sum', 'mean'])
print("\
Transaction Value Analysis:")
print(transaction_value_analysis)
```

```python
# Plot segment distribution
plt.figure(figsize=(10, 6))
sns.countplot(data=user_segments, x='segment')
plt.title('Distribution of Customer Segments')
plt.xticks(rotation=45)
plt.show()
```

Inactive users are high in number

```python
print("Transaction Amount Distribution:")
print(Cleaned_recharge_data['Amount'].describe())
```

```python
custom_bins = [0, 100, 500, 1000, float('inf')]
custom_labels = ['Low', 'Medium', 'High', 'Premium']

# Add transaction segments
Cleaned_recharge_data['transaction_segment'] = pd.cut(Cleaned_recharge_data['Amount'], bins=custom_bins,
                                               labels=custom_labels)

# Analyze revenue by segment
revenue_by_segment = Cleaned_recharge_data.groupby('transaction_segment').agg({
    'Amount': ['count', 'sum', 'mean'],'Uid': 'nunique'}).round(2)

revenue_by_segment.columns = ['Transaction_Count', 'Total_Revenue', 'Avg_Transaction_Value', 'Unique_Users']
print("Revenue Analysis by Transaction Segment:")
print(revenue_by_segment)
```

```python
# Visualize revenue distribution
plt.figure(figsize=(12, 6))
sns.barplot(data=Cleaned_recharge_data, x='transaction_segment', y='Amount')
plt.title('Average Transaction Value by Segment')
plt.xlabel('Segment')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.show()
```
